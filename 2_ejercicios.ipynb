```python
import numpy as np
```

```python
# ejercicio 1
a = np.array([1, 2, 3, 4])
if np.all(a != 0):
    print("Ninguno de los elemenmtos es 0")
else:
    print("hay al menos un 0 en el arreglo")
```

```python
# ejercicio 2
a = np.array([0, 0, 2, 0])
if np.all(a == 0):
    print("no hay elementos diferentes de 0")
else:
    print("Hay al menos un elemento diferente de 0")
```

```python
# ejercicio 3
a = np.array([1, 2, np.nan, np.inf])
np.isnan(a).any() or np.isinf(a).any()
print(f"NaN encontado en: {np.isnan(a)}")
print(f"Inf encontado en: {np.isinf(a)}")
```

```python
# ejercicio 4
a = np.array([1, 2, np.inf])
np.isinf(a).any()
print(f"Inf encontado en: {np.isinf(a)}")
```

```python
# ejercicio 5
a = np.array([1, np.nan, 3])
np.isnan(a).any()
print(f"NaN encontado en: {np.isnan(a)}")
```

```python
# ejercicio 6
a = np.array([1, 3, 5])
b = np.array([2, 3, 4])

resultado = np.where(a > b, "mayor", np.where(a == b, "igual", "menor"))
print(resultado)
```

```python
# ejercicio 7
a = np.array([1, 7, 13, 105])


print(f"Arreglo {a}")
print(f"Memoria ocupada: {a.nbytes} bytes")
print(f"Tipo de dato: {a.dtype}")
print(f"NUmero de elementos: {a.size}")
print(f"Bytes por elemento: {a.itemsize}")
```

```python
# ejecicio 8
ceros = np.zeros(10)
unos = np.ones(10)
cincos = np.full(10, 5)
print(ceros)
print(unos)
print(cincos)

```
